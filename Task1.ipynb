```python

top_counts = df['Continent'].value_counts().head(5)  

# Plot bar chart
plt.figure(figsize=(8,5))
top_counts.plot(kind='bar', color='skyblue')
plt.xlabel("Continent")
plt.ylabel("Number of Countries")
plt.title("Top 5 Continents by Number of Countries")
plt.show()
```
